```python
df = df[df['Pickup point']=='Airport']
```

```python
df
```

```python
df.dropna(axis = 0,how = "any",inplace=True)
```

```python
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])
df['Request timestamp'].dt.time
```

```python
df.sort_values(by=['Request timestamp'], inplace=True)
```

```python
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])

# calculate the time difference between consecutive rows
df['time_diff'] = df['Request timestamp'].diff()

df
```

```python
df[df['time_diff'] > pd.Timedelta(minutes=5)]
```

```python
# prompt: find the difference which is greater than 5 mins from time_diff column

df[df['time_diff'] > pd.Timedelta(minutes=5)].shape[0]
```

```python
df.sort_values(by=['Drop timestamp'])
```

```python
df['less_than5'] = df[df['time_diff'].apply(lambda x:x['time_diff'] > 5)]
```

```python
# call datetime.strptime to convert
# it into datetime datatype
datetime_obj = datetime.strptime(['Request timestamp'].str(),
                                 "%d%b%Y%H%M%S")

# It will print the datetime object
print(datetime_obj)

# extract the time from datetime_obj
time = datetime_obj.time()


# it will print time that
# we have extracted from datetime obj
print(time)
```

```python
df[df['Pickup point'] == 'Airport'].groupby(['Driver id'])['Request timestamp'].diff() > 300
```

```python
df = df[df['Request timestamp'].diff().total_seconds() > 300]

# Count the number of rows remaining.
count = df.shape[0]

# Print the number of drivers who didn't meet any other drivers at the airport.
print(count)
```
